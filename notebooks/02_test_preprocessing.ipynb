```python
import sys
sys.path.append('/app')

from src.data.preprocessing import preprocess_image
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

print("✓ Módulos cargados")
```

```python
# Cargar dataset
TRAIN_PATH = Path('/app/data/raw/Training_Set/Training_Set')
train_df = pd.read_csv(TRAIN_PATH / 'RFMiD_Training_Labels.csv')

# Tomar 1 imagen de cada clase
img_0 = train_df[train_df['Disease_Risk']==0].sample(1)['ID'].values[0]
img_1 = train_df[train_df['Disease_Risk']==1].sample(1)['ID'].values[0]

# Paths
path_0 = TRAIN_PATH / 'Training' / f"{img_0}.png"
path_1 = TRAIN_PATH / 'Training' / f"{img_1}.png"

# Preprocesar
processed_0 = preprocess_image(path_0)
processed_1 = preprocess_image(path_1)

# Visualizar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(processed_0)
axes[0].set_title(f"Sano (ID: {img_0})")
axes[0].axis('off')

axes[1].imshow(processed_1)
axes[1].set_title(f"Enfermo (ID: {img_1})")
axes[1].axis('off')

plt.tight_layout()
plt.show()

print(f"✓ Shape: {processed_0.shape}")
print(f"✓ Rango: [{processed_0.min():.3f}, {processed_0.max():.3f}]")
```
